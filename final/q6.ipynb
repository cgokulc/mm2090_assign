```python
p1=plot(eqn, xmin=0, xmax=25, ymin=0, ymax=50, linestyle='--', color='green', legend_label='Function') #plot of function
```

```python
r=list()
for i in np.arange(min_x,max_x,(max_x-min_x)/no_of_tangents):
    r.append(i)
```

```python
g=Graphics()
for i in r:
    temp_eqn=deqn.substitute(x=i)*x+(eqn.substitute(x=i)-deqn.substitute(x=i)*i)
    g+=plot(temp_eqn,xmin=0, xmax=25, ymin=0, ymax=50, color='yellow') #plot of tangents
g.save_image("q6_tangents.eps")
g+=p1
g.legend((20,5))
g.axes_labels(("x $\\longrightarrow$","y $\\uparrow$"))
g.show()
g.save_image("q6.eps")
```

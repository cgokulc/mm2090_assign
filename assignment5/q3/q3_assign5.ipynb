```python
eq=x**6-x**5-5*x+1
```

```python
roots=eq.complex_roots()
```

```python
rc=0
r=0
coef=list(0 for _ in range(ceil(eq.exponents()[-1]+1)))
for i in range(len(eq.coefficients())):
    coef[eq.exponents()[i]]=eq.coefficients()[i]
for i in coef[0:-3]:
    rc+=abs(i)
d=abs(coef[-2])
dr=abs(rc-1)
if(d<=dr):
    c=[-1*coef[-2],0][max(1,rc)==1]
    r=max(1,rc)
else:
    if(-1*coef[-2]+rc>1):
        c=(-1*coef[-2]-rc+1)/2.0
    else:
        c=(-1*coef[-2]+rc-1)/2.0
    r=(1+rc+d)/2.0
gr=Graphics()
gr.fontsize(8)
gr+=circle((-1*coef[-2],0),rc,color='red')
gr+=circle((0,0),1,color='red')
gr+=circle((c,0),r)
gr+=list_plot(roots,color='red',marker='o')
gr.axes_labels(['real $\\longrightarrow$','imaginary $\\uparrow$'])
gr.show()
```
